```python
def combine_features(row):
    return row['Personality']+" "+row['Passions']+" "+row['Location']+" "+row['Education']+" "+row['Religion']+" "+row['Characteristics_friend']+" "+row['Perfect_match']+" "+row['Hobbies']+" "+row['Imagination']+" "+row['Artistic']+" "+row['Sociable']+" "+row['Reserved']+" "+row['Lazy']+" "+row['Nervous']+" "+row['Trusting']+" "+row['Thorough']+" "+row['Relaxed']
df['soup'] = df.apply(combine_features, axis=1)    
```

```python
#Import TfIdfVectorizer from scikit-learn
from sklearn.feature_extraction.text import TfidfVectorizer


text = df['soup'].values.astype('U') #Taking the text from the df and convert it to Unicode
tfidf = TfidfVectorizer(stop_words='english') #Remove all stopwords

tfidf = tfidf.fit(text) #We fit the model with the words from the text
tfidf


feature_names = tfidf.get_feature_names() #Get the words from the vocabulary
#feature_names
print(f"There are {len(feature_names)} words in the vocabulary. A selection: {feature_names[42:62]}")


#Construct the required TF-IDF matrix by fitting and transforming the data
tfidf_matrix = tfidf.fit_transform(df['soup'])

#Output the shape of tfidf_matrix
tfidf_matrix.shape
```

```python
# Import linear_kernel
from sklearn.metrics.pairwise import linear_kernel

# Compute the cosine similarity matrix
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)
```

```python
df = df.reset_index()
indices = pd.Series(df.index, index=df['User']).drop_duplicates()
```

```python
# Function that takes in a user as input and outputs users
def get_recommendations(User, cosine_sim=cosine_sim):
    # Get the index of the user that matches another user
    idx = indices[User]

    # Get the pairwsie similarity scores 
    sim_scores = list(enumerate(cosine_sim[idx]))

    # Sort the users based on the similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get the scores of the 10 most similar users
    sim_scores = sim_scores[1:11]

    # Get the user indices
    match_indices = [i[0] for i in sim_scores]

    # Return the top 10 most similar users
    return df['User'].iloc[match_indices]
```
