# Recommendation system 2 (text-based)

```python
#Merge two columns because of the similarity in answers 
df['Interests'] = df['Passions']+ ' ' +df['Hobbies']

del df['Passions']
del df['Hobbies']

# Drop duplicate words
df['Interests'] = df['Interests'].str.replace(r'\b(\w+)(\s+\1)+\b', r'\1')
```

```python
#Merge two columns because of the similarity in answer from every user
df['Perfect_match'] = df['Perfect_match']+ ' ' +df['Characteristics_friend']

del df['Characteristics_friend']

# Drop duplicate words
df['Perfect_match'] = df['Perfect_match'].str.replace(r'\b(\w+)(\s+\1)+\b', r'\1')
```

```python
#Merging all columns that contain user preferences and interests and traits
df['Preference'] = df['Interests']+ ' ' +df['Perfect_match']+ ' ' +df['Personality']

df.head()
```

```python
#transform everything to lower case
df['Preference'] = df['Preference'].apply(lambda x:x.lower())
df.head()
```

```python
#Import TfIdfVectorizer from scikit-learn
from sklearn.feature_extraction.text import TfidfVectorizer

text = df['Preference'].values.astype('U') #Converting text from the df and convert it to Unicode
tfidf = TfidfVectorizer(stop_words='english') #Remove stopwords

tfidf = tfidf.fit(text) #We fit the model with the words from the preference text
tfidf

feature_names = tfidf.get_feature_names() #Get the words from the vocabulary
#feature_names
print(f"There are {len(feature_names)} words in the vocabulary. A selection: {feature_names[42:62]}")


#Construct the required TF-IDF matrix by fitting and transforming the data
tfidf_matrix = tfidf.fit_transform(df['Preference'])

#Output the shape of tfidf_matrix
tfidf_matrix.shape
```

```python
# Import linear_kernel
from sklearn.metrics.pairwise import linear_kernel

# Compute the cosine similarity matrix
cosine_sim = linear_kernel(tfidf_matrix, tfidf_matrix)

cosine_sim 
```

```python
#Add index instead of names to preserve respondents privacy
df['User'] = np.arange(len(df))
```

```python
#create df of the similarity scores
df = pd.DataFrame(cosine_sim, columns=df.index, index=df['User']).reset_index()
df.head()
```

```python
#Merge similarity dataframes to one with al similarity scores
df = pd.merge(df, df_num_cat, on="User")

#Drop the user column *this columns does not need to be used for making similarity scores*
df1 = df.drop(columns=['User'])
df1 = pd.DataFrame(df1)
df.info()
```

```python
df = df.reset_index()
indices = pd.Series(df.index, index=df['User']).drop_duplicates()
```

```python
# Function that takes in user as input and outputs most similar users
def get_recommendations(User, df1=df1):
    # Get the index of the users that matches the df
    idx = indices[User]

    # Get the pairwsie similarity scores of all users with df
    sim_scores = list(enumerate(df1[idx])) 
    
    # Sort the user based on the similarity scores
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)

    # Get the scores of the 10 most similar users
    sim_scores = sim_scores[1:12]

    # Get the user indices
    match_indices = [i[0] for i in sim_scores]

    # Return the top 10 most similar users
    return df['User'].iloc[match_indices]
```
